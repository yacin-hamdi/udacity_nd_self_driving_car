```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 
```

```python
image = mpimg.imread("warped-example.jpg")
plt.imshow(image)
print(image.shape)
```

```python
def hist(img):
    bottom_half = img[img.shape[0]//2:,:]
    sums = 0
    histogram = np.sum(bottom_half, axis=0)
    return histogram
```

```python
hist_image = hist(image)
plt.plot(hist_image)
```
