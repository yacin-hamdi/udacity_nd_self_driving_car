```python
import cv2 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import numpy as np 
```

```python
def gradient(img, orient='x', min_thres=0, max_thres=255):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0) if orient == 'x' else cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    abs_sobel = np.absolute(sobel)
    scal_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    binary = np.zeros_like(scal_sobel)
    binary[(scal_sobel >= min_thres) & (scal_sobel <= max_thres)] = 1
    return binary
```

```python
image = mpimg.imread("signs_vehicles_xygrad.png")
gradient_image = gradient(image,orient='x', min_thres=10, max_thres=100)
plt.imshow(gradient_image, cmap='gray')
```
