```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image = mpimg.imread("test4.jpg")
plt.imshow(image)
```

```python
R = image[:, :, 0]
G = image[:, :, 1]
B = image[:, :, 2]
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,12))
ax1.imshow(R, cmap="gray")
ax1.set_title("R")
ax2.imshow(G, cmap="gray")
ax2.set_title("G")
ax3.imshow(B, cmap="gray")
ax3.set_title("B")
```

```python
r_thresholds = (200, 255)
g_thresholds = (200, 255)
b_thresholds = (200, 255)

def thresholds(img, thresholds=(10,100)):
    binary = np.zeros_like(img)
    binary[(img >= thresholds[0])&(img <= thresholds[1])] = 1
    return binary

    
```

```python
r_binary = thresholds(R, thresholds=r_thresholds)
g_binary = thresholds(G, thresholds=g_thresholds)
b_binary= thresholds(B, thresholds=b_thresholds)
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,12))
ax1.imshow(r_binary, cmap='gray')
ax1.set_title('R')
ax2.imshow(g_binary, cmap='gray')
ax2.set_title('G')
ax3.imshow(b_binary, cmap='gray')
ax3.set_title('B')
```

```python
hls_image = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls_image[:, :, 0]
L = hls_image[:, :, 1]
S = hls_image[:, :, 2]

f,(ax1, ax2, ax3) = plt.subplots(1,3,figsize=(12,12))
ax1.imshow(H, cmap="gray")
ax1.set_title("H")
ax2.imshow(L, cmap="gray")
ax2.set_title("L")
ax3.imshow(S, cmap="gray")
ax3.set_title("S")
```

```python
h_thresholds = (15, 30)
l_thresholds = (150, 200)
s_thresholds = (90, 255)

h_binary = thresholds(H, h_thresholds)
l_binary = thresholds(L, l_thresholds)
s_binary = thresholds(S, s_thresholds)
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 12))
ax1.imshow(h_binary, cmap="gray")
ax1.set_title("H")
ax2.imshow(l_binary, cmap="gray")
ax2.set_title("L")
ax3.imshow(s_binary, cmap="gray")
ax3.set_title("S")
```
