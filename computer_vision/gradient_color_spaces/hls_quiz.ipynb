```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def hls_threshold(image, thresholds=(0, 255)):
    hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    S = hls[:, :, 2]
    binary = np.zeros_like(S)
    binary[(S > thresholds[0]) & (S <= thresholds[1])] = 1
    return binary
```

```python
thresholds = (90, 255)
image = mpimg.imread("test6.jpg")
s_binary = hls_threshold(image, thresholds)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 12))
ax1.imshow(image)
ax1.set_title("image")
ax2.imshow(s_binary, cmap="gray")
ax2.set_title("theshold image")
```
