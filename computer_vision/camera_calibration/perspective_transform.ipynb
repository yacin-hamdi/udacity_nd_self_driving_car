```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
image_name = "download.jpg"
image = cv2.imread(image_name)
plt.imshow(image)
plt.plot(175, 16, '.')
plt.plot(187, 58, '.')
plt.plot(51, 105, '.')
plt.plot(60, 62, '.')
```

```python
def warm(img):
    img_size = (img.shape[0], img.shape[1])
    src = np.float32([
        [175, 16], 
        [187, 58], 
        [51, 105], 
        [60, 62]])
    
    dst = np.float32([
        [175, 62], 
        [175, 105], 
        [51, 105], 
        [51, 62]])
    
    
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    return warped
```

```python
warped = warm(image)
plt.imshow(warped)
```
