```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
image_name = "exit-ramp.jpg"
image = mpimg.imread(image_name)
plt.imshow(image)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap="gray")
```

```python
kernel_size = 7
image_blur = cv2.GaussianBlur(image_gray, (kernel_size, kernel_size), 0)
plt.imshow(image_blur, cmap="gray")
```

```python
low_threshold = 50
high_threshold = 150
image_canny = cv2.Canny(image_blur, low_threshold, high_threshold)
plt.imshow(image_canny, cmap="Greys_r")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np


image_name = "exit-ramp.jpg"
image = mpimg.imread(image_name)
plt.imshow(image)
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap="gray")

gaussian_image = cv2.GaussianBlur(image_gray, (5, 5), 0)
plt.imshow(gaussian_image, cmap="gray")

low_threshold = 50
high_threshold = 150
canny_image = cv2.Canny(gaussian_image, low_threshold, high_threshold)
plt.imshow(canny_image, cmap="Greys_r")
```
