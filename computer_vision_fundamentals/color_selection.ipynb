```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image_name = "test.jpg"
street = mpimg.imread(image_name)
```

```python
lane_select = np.copy(street)
red_threshold = 200
green_threshold = 200
blue_threshold = 200
rgb_threshold = [red_threshold, green_threshold, blue_threshold]
```

```python
thresholds = (street[:,:,0] < rgb_threshold[0])|(street[:,:,1] < rgb_threshold[1])|(street[:,:,2] < rgb_threshold[2])
lane_select[thresholds] = [0,0,0]
```

```python
plt.imshow(lane_select)
```

```python
plt.imsave("lane_select.jpg", lane_select)
```
